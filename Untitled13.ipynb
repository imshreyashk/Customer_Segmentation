```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
import datetime as dt
```

```python
df_2010_11 = pd.read_csv('/content/Year 2010-2011.csv', encoding='unicode_escape')
df_2009_10 = pd.read_csv('/content/Year 2009-2010.csv', encoding='unicode_escape')

```

```python
df = pd.concat([df_2009_10, df_2009_10], ignore_index=True)
```

```python
df.info()
```

```python
df.dropna(subset=['Customer ID'], inplace=True)
```

```python
df = df[df['Invoice'].str.contains('C', na=False)]
```

```python
data = df[(df['Quantity'] >0) & (df['Price'] >0)]
```

```python
df['TotalSum'] = df['Quantity'] * df['Price']
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
df['Month'] = df['InvoiceDate'].dt.month
```

```python
df['Month_Year'] = df['InvoiceDate'].dt.to_period('M')
```

```python
# Which countries have the most customers

top_countries = df.groupby('Country')['Customer ID'].nunique().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries, y=top_countries.index, palette='viridis')
plt.title('Top 10 Counties by Customer Count')
plt.xlabel('Number of Unique Customer')
plt.show()
```

```python
#What are the peak hours for shopping--
peak_hours = df.groupby('Hour')['InvoiceDate'].nunique()
plt.figure(figsize=(10, 6))
sns.lineplot(x=peak_hours.index, y=peak_hours.values, marker='o', color='blue')
plt.title('Peak Hours for Shopping')
plt.xticks(range(6,21))
plt.grid(True)
plt.show()
```

```python
#Is there seasonality (Monthly Sales)--

monthly_sales = df.groupby('Month')['TotalSum'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='viridis')
plt.title('Total Revenue by Moneth')
plt.xlabel('Month')
plt.ylabel('Revene')
plt.show()
```

```python
import datetime as dt

# Set the reference date for 'Recency'
latest_date = df['InvoiceDate'].max()
# If InvoiceDate is a string, convert it first:
# df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
snapshot_date = latest_date + dt.timedelta(days=1)

# Group by customer
rfm = df.groupby('Customer ID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days, # Recency
    'Invoice': 'nunique',                                 # Frequency
    'TotalSum': 'sum'                                       # Monetary
})

# Rename for clarity
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'Invoice': 'Frequency',
    'TotalSum': 'Monetary'
}, inplace=True)

print(rfm.describe())
```

```python
from sklearn.cluster import KMeans

#1. Normalize the data (K-Means needs this!)
# We use log to handel the high 'skew' (whales vs small fish)
# The 'Monetary' column currently contains negative values, which causes NaN when applying np.log(x+1).
# To proceed with clustering while acknowledging the negative values (which might indicate cancellations),
# we will take the absolute value of 'Monetary' before applying the log transformation.
# This changes the interpretation of 'Monetary' to the magnitude of the transaction, rather than actual spending.
# A more appropriate long-term fix would involve re-evaluating the 'df' DataFrame
# to include only actual purchases (positive quantities/prices) before RFM calculation.
rfm_log = rfm.copy()
rfm_log['Recency'] = np.log(rfm_log['Recency'] + 1)
rfm_log['Frequency'] = np.log(rfm_log['Frequency'] + 1)
rfm_log['Monetary'] = np.log(np.abs(rfm_log['Monetary']) + 1)

#2 scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_log)

#3. Find the Elbow
sse = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto') # Added n_init='auto' to suppress warning
  kmeans.fit(rfm_scaled)
  sse.append(kmeans.inertia_)

#4. Plot
plt.figure(figsize=(10,6))
plt.plot(range(1,11), sse, marker='o')
plt.title('The elbow method')
plt.xlabel('Number of cluster')
plt.ylabel('Inertia (SSE)')
plt.show()
```

```python
from sklearn.cluster import KMeans

#1. Initialize and fit the model with 4 cluster
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, random_state=42)
cluster_labels = kmeans.fit_predict(rfm_scaled)

#2. Assign these labels back to our RFM dataframe
rfm['Cluster'] = cluster_labels

#3. Calculate the average values for each cluster to see what they represent
df_clusters = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(0)

print(df_clusters)

plt.figure(figsize=(10, 8))
rfm['Cluster'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('viridis'))
plt.title('Customer Segment Distribution')
plt.ylabel('') # Hide the y-label for beauty
plt.show()
```

```python
# Prepare data for Snake Plot
# Create a DataFrame from the scaled RFM data, preserving original column names and index
rfm_normalized = pd.DataFrame(rfm_scaled, index=rfm.index, columns=rfm.columns[:3])

rfm_normalized['Cluster'] = cluster_labels
rfm_melt = pd.melt(rfm_normalized.reset_index(),
                    id_vars=['Customer ID', 'Cluster'], # Corrected 'CustomerID' to 'Customer ID'
                    value_vars=['Recency', 'Frequency', 'Monetary'],
                    var_name='Attribute',
                    value_name='Value')

plt.figure(figsize=(10, 6))
sns.lineplot(x='Attribute', y='Value', hue='Cluster', data=rfm_melt, palette='Set1')
plt.title('Snake Plot of RFM Clusters')
plt.show()
```
